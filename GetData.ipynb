```python
import requests
from bs4 import BeautifulSoup
```

```python
lst = []
for i in range(1,100):
    resp = requests.get('https://mogi.vn/mua-can-ho-chung-cu?cp='+str(i))
    soup = BeautifulSoup(resp.content,'html.parser')
    links = soup.find_all('a',class_='link-overlay')
    for link in links:
    	lst.append(link['href'])
```

```python
X = []
Y = []
error = []
```

```python
for i in lst:
	test = requests.get(i)
	soup1 = BeautifulSoup(test.content,'html.parser')
	info = soup1.find_all(class_='info-attr')
	if len(info) != 6: error.append(i)
	for j in info:
		X.append((j.text.strip())[j.text.strip().find("\n")+1::])
X
```
